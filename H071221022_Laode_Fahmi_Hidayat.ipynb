```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi aktivasi perceptron
def aktivasi(x, w, b):
    z = np.dot(x, w) + b
    return 1 if z >= 0 else 0

# Input dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Output yang diharapkan
y = np.array([0, 1, 1, 1])

# Inisialisasi bobot dan bias tanpa seed tetap
w = np.random.rand(2) - 0.5
b = np.random.rand() - 0.5

# Print initial random weights and bias
print("Bobot acak awal:", w)
print("Bias acak awal:", b)

# Pelatihan perceptron
learning_rate = 0.1
epochs = 7
errors = []

for epoch in range(epochs):
    total_error = 0
    print(f"\nEpoch {epoch + 1}/{epochs}")
    print("="*20)
    for i in range(len(X)):
        x = X[i]
        target = y[i]
        initial_w = w.copy()
        initial_b = b
        net_input = np.dot(x, w) + b
        output = aktivasi(x, w, b)

        # Update weights and bias for each data point
        if output != target or epoch >= 3:  # Conditional update
            w += learning_rate * (target - output) * x
            b += learning_rate * (target - output)

        # Calculate the squared error
        total_error += (target - output)**2

        print(f"Input: {x}, Target: {target}, Initial Weight: {initial_w}, Net: {net_input}, Output: {output}, Error: {target - output}, Updated Weight: {w}, Updated Bias: {b}")

    errors.append(total_error)
    print(f"Epoch {epoch + 1} Error: {total_error}")

# Hasil Prediksi
print("\nHasil Prediksi:")
for x in X:
    output = aktivasi(x, w, b)
    print(x, "->", output)

# Plot error
plt.figure()
plt.plot(range(1, epochs + 1), errors, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Perceptron Learning Curve')
plt.grid(True)
plt.show()

print("Bobot akhir:", w)
print("Bias akhir:", b)
```
